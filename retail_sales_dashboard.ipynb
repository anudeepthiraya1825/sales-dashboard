```python
!pip install pandas matplotlib seaborn

# ðŸ“Š Step 2: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Superstore.csv")
print(" Dataset Loaded Successfully")
print(df.head())

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month_name()

print("\nðŸ”Ž Dataset Info:")
print(df.info())

category_sales = df.groupby("Category")["Sales"].sum().reset_index()

plt.figure(figsize=(6,4))
sns.barplot(data=category_sales, x="Category", y="Sales", palette="viridis")
plt.title("Sales by Category")
plt.ylabel("Total Sales ($)")
plt.xlabel("Category")
plt.savefig("sales_by_category.png")
plt.show()


monthly_sales = df.groupby(df['Order Date'].dt.to_period("M"))["Sales"].sum()

monthly_sales.plot(figsize=(12,5), marker='o', title="Monthly Sales Trend")
plt.ylabel("Sales ($)")
plt.xlabel("Month")
plt.grid(True)
plt.savefig("monthly_sales.png")
plt.show()

region_profit = df.groupby("Region")["Profit"].sum().reset_index()

plt.figure(figsize=(6,4))
sns.barplot(data=region_profit, x="Region", y="Profit", palette="coolwarm")
plt.title("Profit by Region")
plt.savefig("profit_by_region.png")
plt.show()

plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="Sales", y="Profit", hue="Category", style="Segment", s=100)
plt.title("Sales vs Profit by Category & Segment")
plt.savefig("sales_vs_profit.png")
plt.show()


df.to_csv("Superstore_Cleaned.csv", index=False)
print("Cleaned dataset saved as Superstore_Cleaned.csv")
```
